```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
path = '../save/light_test_MA_1_MP_0.5_1e-05_1_2311272202'

pred = np.load(os.path.join(path, 'test_y_preds.npy'))
pred_15 = np.load(os.path.join(path, 'test_y_preds_15.npy'))

Q_pred = pred[:, 0, :, :]
K_pred = pred[:, 1, :, :]
Q_pred_15 = pred_15[:, 0, :, :]
K_pred_15 = pred_15[:, 1, :, :]

Q_pred = np.squeeze(Q_pred)
K_pred = np.squeeze(K_pred)
Q_pred_15 = np.squeeze(Q_pred_15)
K_pred_15 = np.squeeze(K_pred_15)

# 计算Q均值
Q_pred_15_mean = np.mean(Q_pred_15, axis=1)
Q_pred_15_mean = np.mean(Q_pred_15_mean)
Q_pred_mean = np.mean(Q_pred, axis=1)
Q_pred_mean = np.mean(Q_pred_mean)
print(Q_pred_15_mean, Q_pred_mean)

# 计算K均值
K_pred_15_mean = np.mean(K_pred_15, axis=1)
K_pred_15_mean = np.mean(K_pred_15_mean)
K_pred_mean = np.mean(K_pred, axis=1)
K_pred_mean = np.mean(K_pred_mean)
print(K_pred_15_mean, K_pred_mean)


# 计算Q_pred_15对Q_pred的mape
mape = np.mean(np.abs(Q_pred_15 - Q_pred) / Q_pred)*100
mape = np.mean(np.abs(K_pred_15 - K_pred) / K_pred)*100
print(mape)


```
